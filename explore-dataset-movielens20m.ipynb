## Pruning tags

```python
tag_list = list(tag_df.tag.unique())
len(tag_list)
```

```python
tag_list2 =[]
for tag in tag_list:
    if isinstance(tag, str):
        tag_words = tag.split(" ")
        use = True
        if len(tag_words) > 2:
            # Remove tags with more than two words
            use = False
        if not tag.isalnum():
            # Remove tags with special characters
            use = False
#         if len(tag_words) == 1 and len(tag_words[0]) == 1:
#             # remove tags with a single alphabet
#             use = False
        if use:
            tag_list2.append(tag)
# Create unique tag set
tag_set = set(tag_list2)
len(tag_set)
```

### Create tag clusters so that tags with similar semantic meanings can be grouped together

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import torch
from tqdm import tqdm
```

```python
# define your word list
words_list=list(tag_set)
# Embed the entire corpus for semantic search
corpus = words_list
# Select model for semantic similarity search
model = SentenceTransformer('stsb-roberta-large')
# encode corpus to get corpus embeddings
corpus_embeddings = model.encode(corpus, convert_to_tensor=True)
```

```python
def create_cluster(corpus, corpus_embeddings, threshold = 0.75):

    tag_to_clusterid = {}
    clusterid_to_tag = {}
    
#     # Embed the entire corpus for semantic search
#     corpus = words_list
#     # Select model for semantic similarity search
#     model = SentenceTransformer('stsb-roberta-large')
#     # encode corpus to get corpus embeddings
#     corpus_embeddings = model.encode(corpus, convert_to_tensor=True)
    
    # top_k results to return
    top_k=len(corpus)
    
    for index in tqdm(range(len(corpus))):
        ref_tag = corpus[index]
        id1 = index
        if ref_tag not in tag_to_clusterid:
            # New tag registered
            tag_to_clusterid[ref_tag] = id1
            clusterid_to_tag[id1] = set()
            clusterid_to_tag[id1].add(ref_tag) 
        else:
            # tag already present
            id1 = tag_to_clusterid[ref_tag]
            
        # iterate through each embedding as a reference
        sentence_embedding = corpus_embeddings[index]
        
        # compute similarity scores of the sentence with the corpus
        cos_scores = util.pytorch_cos_sim(sentence_embedding, corpus_embeddings)[0]
        # Sort the results in decreasing order and get the first top_k
        top_results = np.argpartition(-cos_scores, range(top_k))[0:top_k]
        for idx in top_results[0:top_k]:
            if cos_scores[idx] < threshold:
                # Make sure to club tags with better cosine similarity scores using a threshold
                break
            tag = corpus[idx]
            if tag not in tag_to_clusterid:
                # New tag registered
                tag_to_clusterid[tag] = id1
                clusterid_to_tag[id1].add(tag)
            else:
                id2 = tag_to_clusterid[tag]
                if id1 != id2:
                    # Combine the two groups with same tag but different ids
                    tags = clusterid_to_tag[id2]
                    for t in tags:
                        tag_to_clusterid[t] = id1
                        clusterid_to_tag[id1].add(t)
                    
                    del clusterid_to_tag[id2]
#                     print(tag_to_clusterid)
                    
#                     # tag already present in dictionary but cannot have two different ids
#                     raise Exception(f"There cannot be two different id {id1}, {id2} for {tag}")
                
    return tag_to_clusterid, clusterid_to_tag
```

```python

tag_to_clusterid, clusterid_to_tag = create_cluster(corpus, corpus_embeddings, threshold = 0.8)
```

## Filter only releavant tags

```python
def filter_tags(x):
    out = False
    if x in tag_set:
        out = True
    return out

def apply_cluster_tag(x):
    new_tagid = -1
    if x in tag_to_clusterid:
        new_tagid = tag_to_clusterid[x]
    return new_tagid

# Filter out relevant tags
tag_df['relevant_tags'] = tag_df.apply(lambda row: filter_tags(row['tag']),axis=1)
tag_df['new_tagid'] = tag_df.apply(lambda row: apply_cluster_tag(row['tag']),axis=1)
```

```python
tag_df1 = tag_df[tag_df.relevant_tags == True]
tag_df1 = tag_df[tag_df.new_tagid != -1]
```

## Simplify Genres in movie dataset

```python
movie.shape
```

```python
genre_list = list(movie.genres.unique())
```

```python
genre_set = set()
for words in genre_list:
    wordlist = words.split("|")
    for word in wordlist:
        genre_set.add(word)
```

```python
new_genre_list = list(genre_set)
new_genre_list
```

```python
# Create columns for genres
def check_genre(words, c):
    out = False
    wordlist = words.split("|")
    word_set = set(wordlist)
    if c in word_set:
        out = True
    return out

for col in new_genre_list:
    col_name = "genre:" + col
    movie[col_name] = movie.apply(lambda x: check_genre(x['genres'], col), axis=1)
```
